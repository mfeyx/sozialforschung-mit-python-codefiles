```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
data = np.random.random((10, 4))
names = ["f1", "f2", "f3", "f4"]
idx = range(4, 14)
df = pd.DataFrame(data, columns=names, index=idx)
df
```

```python
np.random.seed(42)
cluster_values = np.random.randint(1, 4, 10)
df["cluster"] = cluster_values
df
```

```python
mapping = {
    1: "Sport",
    2: "Wirtschaft",
    3: "Kultur"
}

df["label"] = df.cluster.map(mapping)
df
```

```python
df[(df.label == "Kultur") | (df.cluster == 2)]
```

```python
df[(df.label == "Kultur") & (df.f3 >= 0.3)]
```

```python
def sum_sqrt(array):
    sum_array = sum(array)
    return np.sqrt(sum_array)

df.f1.agg(["min", "max", "mean", "median", sum, sum_sqrt])
```

```python
df.f1.quantile((0.01, 0.25, 0.5, 0.75, 0.99))
```

```python
df["sum_f1_f2"] = df.f1 + df.f2
df
```

```python
df["mean_f1_f2"] = (df.f1 + df.f2) / 2
df
```

```python
df.drop(["sum_f1_f2", "mean_f1_f2"], axis=1, inplace=True)
```

```python
for column in df.columns[0:4]:
    name = f"{column}_mn_diff"
    df[name] = df[column] - df[column].mean()
df
```

```python
def my_func(x, *args):
    for arg in args:
        x *= arg
    return int(x)

df.f1.apply(my_func, args=(3, ))
```
